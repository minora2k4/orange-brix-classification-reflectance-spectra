```python
import os
import pandas as pd

# Đường dẫn tới thư mục chính chứa các thư mục A, B, C, D
main_folder = r'D:\\Cam1'

# Hàm kiểm tra các giá trị trống trong file phổ
def check_missing_values(file_path):
    try:
        # Đọc dữ liệu từ file
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
            lines = f.readlines()

        # Kiểm tra xem có chuỗi ">>>>>Begin Spectral Data<<<<<" không
        if '>>>>>Begin Spectral Data<<<<<\n' in lines:
            start_index = lines.index('>>>>>Begin Spectral Data<<<<<\n') + 1
            
            # Tách dữ liệu ở các dòng tiếp theo
            spectral_data = []
            for line in lines[start_index:]:
                line = line.strip()
                if line:
                    parts = line.split()
                    if len(parts) == 2:
                        spectral_data.append(parts)
            
            # Chuyển dữ liệu thành DataFrame
            if spectral_data:
                data = pd.DataFrame(spectral_data, columns=['Wavelength', 'Intensity'], dtype=float)

                # Kiểm tra xem có giá trị thiếu (NaN) trong bất kỳ cột nào không
                missing_data = data.isnull().sum()

                if missing_data.any():  # Nếu có giá trị trống hoặc NaN
                    print(f"File {file_path} có dữ liệu thiếu:")
                    print(f"Wavelength: {missing_data['Wavelength']} giá trị thiếu")
                    print(f"Intensity: {missing_data['Intensity']} giá trị thiếu")
                else:
                    print(f"File {file_path} không có dữ liệu thiếu.")
            else:
                print(f"[{file_path}] Không tìm thấy dữ liệu phổ sau begin line.")
        else:
            print(f"[{file_path}] Không tìm thấy dòng '>>>>>Begin Spectral Data<<<<<'")
    
    except Exception as e:
        print(f"Không thể kiểm tra file {file_path} do lỗi: {e}")

# Duyệt qua các thư mục A, B, C, D
for folder in os.listdir(main_folder):
    folder_path = os.path.join(main_folder, folder)
    
    if os.path.isdir(folder_path):  # Kiểm tra nếu là thư mục
        # Duyệt qua các thư mục con như A1, A2, ...
        for subfolder in os.listdir(folder_path):
            subfolder_path = os.path.join(folder_path, subfolder)
            
            if os.path.isdir(subfolder_path):  # Kiểm tra nếu là thư mục con
                # Duyệt qua tất cả các file .txt trong thư mục con
                txt_files = [file for file in os.listdir(subfolder_path) if file.endswith(".txt")]
                for file in txt_files:
                    file_path = os.path.join(subfolder_path, file)
                    # Kiểm tra dữ liệu trống trong file
                    check_missing_values(file_path)
```

```python
import os
import re
import random
import shutil
import pandas as pd
import numpy as np
from scipy.signal import savgol_filter

# Hàm chuẩn hóa dữ liệu về [0, 1]
def normalize_data(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))

# Hàm làm mượt phổ sử dụng Savitzky-Golay
def smooth_spectrum(spectrum, window_length=51, polyorder=3):
    if len(spectrum) < window_length:
        # Điều chỉnh window_length sao cho là số lẻ
        window_length = len(spectrum) if len(spectrum) % 2 == 1 else len(spectrum) - 1
    return savgol_filter(spectrum, window_length=window_length, polyorder=polyorder)

def create_brix_folders(root_folder):
    """Tạo các thư mục min_to_6, 6_to_9, 9_to_max trong thư mục gốc"""
    brix_folders = ['min_to_6', '6_to_9', '9_to_max']
    for folder in brix_folders:
        folder_path = os.path.join(root_folder, folder)
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

def move_file_based_on_brix(file_path, label, root_folder):
    """Di chuyển tệp vào thư mục tương ứng dựa trên độ ngọt Brix"""
    if label < 6:
        target_folder = os.path.join(root_folder, 'min_to_6')
    elif 6 <= label < 9:
        target_folder = os.path.join(root_folder, '6_to_9')
    else:
        target_folder = os.path.join(root_folder, '9_to_max')
    
    shutil.move(file_path, os.path.join(target_folder, os.path.basename(file_path)))

def load_and_split_spectra(data_dir):
    """Load spectral data, manually split into train/test sets per fruit, shuffle them, and return X_train, X_test, y_train, y_test, with statistics."""
    X_train, X_test, y_train, y_test = [], [], [], []
    stats = {}

    for category in os.listdir(data_dir):  # Loop over categories A, B, C, D
        category_path = os.path.join(data_dir, category)
        if not os.path.isdir(category_path):
            continue

        for fruit in os.listdir(category_path):  # Loop over fruit folders (e.g., A00, A01,...)
            fruit_path = os.path.join(category_path, fruit)
            if not os.path.isdir(fruit_path):
                continue

            spectra_list = []
            labels = []

            for file in os.listdir(fruit_path):  # Loop over spectral .txt files
                if not file.endswith(".txt"):
                    continue

                try:
                    match = re.search(r"\{([\d.]+)\}", file)
                    if match:
                        label = float(match.group(1))
                    else:
                        continue

                    file_path = os.path.join(fruit_path, file)
                    with open(file_path, 'r') as f:
                        lines = f.readlines()

                    start_index = None
                    for i, line in enumerate(lines):
                        if ">>>>>Begin Spectral Data<<<<<" in line:
                            start_index = i + 2
                            break
                    if start_index is None:
                        continue

                    spectral_data = [line.strip().split() for line in lines[start_index:] if line.strip()]
                    df = pd.DataFrame(spectral_data, columns=["wavelength", "intensity"], dtype=float)
                    norm_intensity = df["intensity"].values / np.max(df["intensity"].values)
                    norm_intensity = smooth_spectrum(norm_intensity, window_length=51, polyorder=3)
                    spectra_list.append(norm_intensity)
                    labels.append(label)

                    # Di chuyển file vào thư mục phù hợp dựa trên Brix
                    move_file_based_on_brix(file_path, label, data_dir)
                    
                except Exception as e:
                    print(f"Error processing {file}: {e}")
                    continue

            # Split per fruit
            if spectra_list:
                combined = list(zip(spectra_list, labels))
                random.shuffle(combined)
                split_idx = int(len(combined) * 0.8)
                train_data, test_data = combined[:split_idx], combined[split_idx:]

                X_train.extend([x for x, _ in train_data])
                y_train.extend([y for _, y in train_data])
                X_test.extend([x for x, _ in test_data])
                y_test.extend([y for _, y in test_data])

                stats[fruit] = {
                    "total": len(combined),
                    "train": len(train_data),
                    "test": len(test_data)
                }

    # Convert to numpy arrays
    X_train = np.array(X_train)
    X_test = np.array(X_test)
    y_train = np.array(y_train)
    y_test = np.array(y_test)

    # Display statistics
    stats_df = pd.DataFrame.from_dict(stats, orient='index')
    print(stats_df)

    return X_train, X_test, y_train, y_test, stats_df

# Create folders for Brix categories
root_folder = r'D:\\Cam1'
create_brix_folders(root_folder)

# Load and split the spectra
X_train, X_test, y_train, y_test, stats_df = load_and_split_spectra(root_folder)

# Now X_train, X_test, y_train, y_test are ready for training any model like SVM, KNN, etc.
```
